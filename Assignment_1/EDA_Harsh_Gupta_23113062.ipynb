```python
df = df.drop(df[df['CoapplicantIncome'] > 15000].index)
df = df.drop(df[df['ApplicantIncome'] > 50000].index)
df.reset_index(drop=True, inplace=True)
df.shape
```

```python
gender_counts = df['Gender'].value_counts()
plt.bar(gender_counts.index, gender_counts, color = sns.color_palette('Spectral'))
plt.title('Gender Distribution')
plt.xlabel('Genders')
plt.ylabel('Counts')
```

# Summary

- The majority of individuals in the dataset are males, and most of them are married. Conversely, most females are unmarried.
- While most people don't have any dependents, men tend to have more than three dependents, indicating potential family responsibilities.
- Income distribution is concentrated between 2000 and 5000, with a significant portion falling within this range. Men, on average, tend to have slightly higher incomes.
- Loan amounts typically range from 50 to 200, with about 70% of individuals not having a loan. Men, on average, have slightly higher loan amounts.
- Approximately 40% of the dataset's population resides in rural areas, and these individuals tend to have higher incomes and loan amounts.
- About three-quarters of individuals are graduates, indicating a relatively educated population.
- Most individuals do not have a self-employed status, suggesting that a majority may be employed by others.
- There are no significant correlations between numeric columns in the dataset, indicating a lack of strong linear relationships.
